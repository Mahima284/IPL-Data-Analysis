```python
matches_clean=matches.copy()
deliviries_clean = deliveries.copy()
matches_clean=matches_clean[matches_clean['winner'].notna()]
print(f"Matches after removing incomplete games: {len(matches_clean)}")
matches_clean['date']=pd.to_datetime(matches_clean['date'])
matches_clean['year']=matches_clean['date'].dt.year
print("\n Cleaned matches shape :",matches_clean.shape)
print("\n Year range :",matches_clean['year'].min(),"to",matches_clean['year'].max())
print("\n Teams in dataset:")
print(sorted(set(matches_clean['team1'].unique().tolist()+matches_clean['team2'].unique().tolist())))
```

cleaned dataset has 1,090 complete IPL matches from 2008 to 2024, with 21 columns of match data. The dataset covers 19 IPL teams including established franchises like CSK and MI, plus newer teams like Lucknow Super Giants and rebranded teams like Royal Challengers Bengaluru. This gives solid data across 16 years to analyze team performance, player stats, and tournament patterns.

BATSMAN ANALYSIS

```python
batsman_stats = deliviries_clean.groupby('batter').agg({
    'batsman_runs' : 'sum',
    'ball':'count'
}).rename(columns={'batsman_runs':'total_runs','ball':'balls_faced'})
batsman_stats['strike_rate']=batsman_stats['total_runs']/batsman_stats['balls_faced']*100
batsman_stats=batsman_stats[batsman_stats['total_runs']>=500]
batsman_stats=batsman_stats.sort_values('strike_rate',ascending=False)
print("Top 10 Batsmen by Strike Rate(min 500 runs):")
print(batsman_stats.head(10))
```

Analysis
This batsman analysis groups all ball-by-ball delivery data by batter name, sums up the total runs each batsman scored, counts the balls they faced, then calculates strike rate (runs per 100 balls). After filtering to include only batsmen with at least 500 runs (to exclude occasional players), the results show the top 10 most aggressive scorers in IPL history. PD Salt leads with a strike rate of 169.6, meaning he scores nearly 170 runs per 100 balls faced—extremely high and rare. AD Russell and GJ Maxwell rank among the top scorers with 2,488 and 2,772 runs respectively, but their strike rates (164 and 150) are still excellent. These batsmen represent the most explosive hitters in IPL, scoring quickly and aggressively regardless of match situation, making them valuable assets in T20 cricket where quick runs are critical

```python
BOWLER ANALYSIS
```

```python
bowler_stats = deliveries_clean.groupby('bowler').agg({
    'total_runs': 'sum',
    'ball': 'count'
}).rename(columns={'total_runs': 'runs_conceded', 'ball': 'balls_bowled'})

bowler_stats['overs_bowled'] = bowler_stats['balls_bowled'] / 6

bowler_stats['economy_rate'] = bowler_stats['runs_conceded'] / bowler_stats['overs_bowled']

bowler_stats = bowler_stats[bowler_stats['balls_bowled'] >= 60]

bowler_stats = bowler_stats.sort_values('economy_rate', ascending=True)

print("Top 10 Bowlers by Economy Rate (min 60 balls bowled):")
print(bowler_stats.head(10))
```

Analysis
This bowler analysis groups deliveries by bowler, calculates total runs conceded and overs bowled, then computes economy rate (runs per over). Lower economy rate means better bowling control. Sohail Tanvir leads with 6.23 runs per over, the most economical bowler. Most top bowlers concede 6.2-6.6 runs per over. Bowlers with more overs (like Kumble with 163 overs) have slightly higher economy rates because their statistics are spread across larger sample sizes

TEAM PERFORMANCE

```python
team_wins_As_team1=matches_clean[matches_clean['winner']== matches_clean['team1']].groupby('team1').size()
team_matches_As_team1=matches_clean['team1'].value_counts()
team_performance=pd.DataFrame({
    'matches_played':team_matches_As_team1,
    'wins': team_wins_As_team1}).fillna(0)


team_performance['win_percentage'] = (team_performance['wins'] / team_performance['matches_played'] * 100).round(2)

team_performance = team_performance.sort_values('win_percentage', ascending=False)


print("Team Performance (Win Percentage):")
print(team_performance)
```

Analysis
This team analysis calculates win percentage for each team based on matches played and wins.This team analysis calculates win percentage for each team based on matches played and wins. Lucknow Super Giants leads with 72.73% win rate, while established franchises like Chennai Super Kings (58.59%) and Mumbai Indians (56.91%) show consistent dominance over many matches. Newer teams have higher percentages but fewer matches, while Pune Warriors has the lowest win rate at 26.09%

```python
TOSS IMPACT
```

```python
toss_analysis = matches_clean.groupby('toss_winner').agg({
    'winner': 'count'
}).rename(columns={'winner': 'toss_wins_total'})

toss_analysis['wins_after_toss'] = matches_clean[matches_clean['winner'] == matches_clean['toss_winner']].groupby('toss_winner').size()

toss_analysis['win_percentage'] = (toss_analysis['wins_after_toss'] / toss_analysis['toss_wins_total'] * 100).round(2)

toss_analysis = toss_analysis.sort_values('win_percentage', ascending=False)

print("Toss Impact on Match Outcome:")
print(toss_analysis)
```

Analysis
This toss analysis shows the win percentage for each team after winning the coin toss. Rising Pune Supergiant wins 83.33% of matches after winning toss, while most teams win 50-83% after toss advantage. Chennai Super Kings wins 61.98% and Mumbai Indians 54.55%. The data indicates winning toss provides a meaningful but not decisive advantage—team strength and strategy matter more, as Pune Warriors wins only 15% despite toss advantage

VISUALIZATIONS

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Plot 1: Top 10 batsmen by strike rate
top_batsmen = batsman_stats.head(10)
axes[0, 0].barh(range(len(top_batsmen)), top_batsmen['strike_rate'])
axes[0, 0].set_yticks(range(len(top_batsmen)))
axes[0, 0].set_yticklabels(top_batsmen.index)
axes[0, 0].set_xlabel('Strike Rate')
axes[0, 0].set_title('Top 10 Batsmen by Strike Rate')
axes[0, 0].invert_yaxis()

# Plot 2: Top 10 bowlers by economy rate
top_bowlers = bowler_stats.head(10)
axes[0, 1].barh(range(len(top_bowlers)), top_bowlers['economy_rate'])
axes[0, 1].set_yticks(range(len(top_bowlers)))
axes[0, 1].set_yticklabels(top_bowlers.index)
axes[0, 1].set_xlabel('Economy Rate')
axes[0, 1].set_title('Top 10 Bowlers by Economy Rate')
axes[0, 1].invert_yaxis()

# Plot 3: Top 10 teams by win percentage
top_teams = team_performance.head(10)
axes[1, 0].bar(range(len(top_teams)), top_teams['win_percentage'])
axes[1, 0].set_xticks(range(len(top_teams)))
axes[1, 0].set_xticklabels(top_teams.index, rotation=45, ha='right')
axes[1, 0].set_ylabel('Win Percentage')
axes[1, 0].set_title('Top 10 Teams by Win Percentage')

# Plot 4: Toss impact
toss_top = toss_analysis.head(10)
axes[1, 1].bar(range(len(toss_top)), toss_top['win_percentage'])
axes[1, 1].set_xticks(range(len(toss_top)))
axes[1, 1].set_xticklabels(toss_top.index, rotation=45, ha='right')
axes[1, 1].set_ylabel('Win % After Winning Toss')
axes[1, 1].set_title('Toss Impact on Match Outcome')

plt.tight_layout()
plt.savefig('ipl_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("Visualization saved as 'ipl_analysis.png'")
```
